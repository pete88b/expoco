# Scoping

## Define project

### Goal

Control the pointer using facial expressions. e.g.
- eyes control pointer position
- blink, cheek puff, head tilt etc could control clicks
- switched on/off with a shrug

### Why

Improve workflows by removing the need to leave the keyboard "home poisition". e.g.
- moving between different lines of code/text
- moving to a specific section of a different window

### Feasibility and Value
- feasibility - aim to complete in < 1 week
    - get pointer position control working
    - show that we can get F1 > 90% for 3 expressions
- value 
    - can we use the above to augment pointer control without being intrusive
        - i.e. can we do some of the "Why" by partially achieving some goal(s)

### Key metrics
